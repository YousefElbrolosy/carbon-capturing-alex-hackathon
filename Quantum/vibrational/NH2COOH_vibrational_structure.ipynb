```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from typing import List, Tuple, Optional, Dict, Any

# Qiskit imports
from qiskit import QuantumCircuit
from qiskit_algorithms.optimizers import SLSQP
from qiskit.circuit import Parameter, ParameterVector
from qiskit.quantum_info import SparsePauliOp, Statevector
from qiskit.primitives import StatevectorEstimator
from qiskit_algorithms.optimizers import COBYLA, SPSA, L_BFGS_B
from qiskit_algorithms.minimum_eigensolvers import VQE
from qiskit.circuit.library import EfficientSU2

class NH2COOH_Qiskit_VQE:
    """
    Variational Quantum Eigensolver for NH2COOH (Carbamic Acid) ground state vibrational energy using Qiskit
    """
    
    def __init__(self, n_qubits: int = 6, reps: int = 2):
        """
        Initialize the Qiskit VQE simulator for NH2COOH
        
        Args:
            n_qubits: Number of qubits for the quantum circuit
            reps: Number of repetitions for the ansatz circuit
        """
        self.n_qubits = n_qubits
        self.reps = reps
        
        # NH2COOH vibrational parameters (estimated from literature and similar molecules)
        # These are typical frequencies for carbamic acid functional groups
        self.omega_nh2_as = 3550    # cm^-1, NH2 asymmetric stretch
        self.omega_nh2_s = 3450     # cm^-1, NH2 symmetric stretch
        self.omega_ch_cooh = 3000   # cm^-1, COOH O-H stretch (broad)
        self.omega_co = 1750        # cm^-1, C=O stretch
        self.omega_cn = 1350        # cm^-1, C-N stretch
        self.omega_nh2_bend = 1600  # cm^-1, NH2 bending
        self.omega_coh_bend = 1300  # cm^-1, C-OH bending
        self.omega_ccn_bend = 800   # cm^-1, C-C-N bending
        
        # Collect all frequencies for easier handling
        self.frequencies = [
            self.omega_nh2_as,    # Mode 0: NH2 antisymmetric stretch
            self.omega_nh2_s,     # Mode 1: NH2 symmetric stretch  
            self.omega_ch_cooh,   # Mode 2: COOH O-H stretch
            self.omega_co,        # Mode 3: C=O stretch
            self.omega_cn,        # Mode 4: C-N stretch
            self.omega_nh2_bend,  # Mode 5: NH2 bending
            # Add more modes if n_qubits > 6
            self.omega_coh_bend,  # Mode 6: C-OH bending
            self.omega_ccn_bend,  # Mode 7: C-C-N bending
        ]
        
        # Extend with lower frequency modes if more qubits needed
        while len(self.frequencies) < n_qubits:
            self.frequencies.append(500 + 200 * (len(self.frequencies) - 8))
        
        # Convert to atomic units (Hartree)
        self.cm_to_hartree = 4.556e-6
        self.frequencies_au = [freq * self.cm_to_hartree for freq in self.frequencies]
        
        # Coupling constants (estimated)
        self.coupling_strong = 0.002   # Strong coupling between NH2 modes and C=O
        self.coupling_medium = 0.001   # Medium coupling between other modes
        self.coupling_weak = 0.0005    # Weak coupling for distant modes
        
        # VQE tracking
        self.energy_history = []
        self.iteration_count = 0
        
    def create_hamiltonian_pauli(self) -> SparsePauliOp:
        """
        Create the vibrational Hamiltonian for NH2COOH as a Pauli operator sum
        
        Returns:
            SparsePauliOp: Hamiltonian in Pauli string representation
        """
        pauli_list = []
        coeffs = []
        
        # Zero-point energy (constant term)
        zpe = 0.5 * sum(self.frequencies_au[:self.n_qubits])
        pauli_list.append('I' * self.n_qubits)
        coeffs.append(zpe)
        
        print(f"Zero-point energy: {zpe:.6f} Hartree ({zpe/self.cm_to_hartree:.2f} cm^-1)")
        
        # Single-mode energies: _i * (I - Z_i)/2
        # This maps |0  0, |1  _i for each qubit
        for i in range(self.n_qubits):
            omega_i = self.frequencies_au[i]
            
            # Identity term: _i/2
            pauli_str = 'I' * self.n_qubits
            pauli_list.append(pauli_str)
            coeffs.append(omega_i / 2)
            
            # Z term: -_i/2 * Z_i
            pauli_str = 'I' * i + 'Z' + 'I' * (self.n_qubits - i - 1)
            pauli_list.append(pauli_str)
            coeffs.append(-omega_i / 2)
        
        # Coupling terms: coupling * Z_i * Z_j with chemistry-informed strengths
        for i in range(self.n_qubits):
            for j in range(i + 1, self.n_qubits):
                # Determine coupling strength based on chemical relationship
                coupling_strength = self._get_coupling_strength(i, j)
                
                if coupling_strength > 0:
                    pauli_str = ['I'] * self.n_qubits
                    pauli_str[i] = 'Z'
                    pauli_str[j] = 'Z'
                    pauli_list.append(''.join(pauli_str))
                    coeffs.append(coupling_strength)
        
        return SparsePauliOp.from_list(list(zip(pauli_list, coeffs)))
    
    def _get_coupling_strength(self, i: int, j: int) -> float:
        """
        Determine coupling strength between modes i and j based on chemical intuition
        
        Args:
            i, j: Mode indices
            
        Returns:
            Coupling strength
        """
        # Mode assignments (based on our frequency list):
        # 0: NH2 antisymmetric stretch
        # 1: NH2 symmetric stretch
        # 2: COOH O-H stretch
        # 3: C=O stretch
        # 4: C-N stretch
        # 5: NH2 bending
        
        # Strong coupling cases
        if (i, j) in [(0, 1), (1, 0)]:  # NH2 stretches
            return self.coupling_strong
        elif (i, j) in [(3, 4), (4, 3)]:  # C=O and C-N stretches
            return self.coupling_strong
        elif (i, j) in [(2, 3), (3, 2)]:  # COOH O-H and C=O
            return self.coupling_strong
        
        # Medium coupling cases
        elif (i <= 1 and j == 5) or (i == 5 and j <= 1):  # NH2 stretches with NH2 bend
            return self.coupling_medium
        elif (i <= 4 and j >= 4) or (i >= 4 and j <= 4):  # Stretch-bend coupling
            return self.coupling_medium
        
        # Weak coupling for other combinations
        else:
            return self.coupling_weak
    
    def create_ansatz_circuit(self) -> QuantumCircuit:
        """
        Create parameterized ansatz circuit using EfficientSU2
        
        Returns:
            QuantumCircuit: Parameterized quantum circuit
        """
        ansatz = EfficientSU2(
            num_qubits=self.n_qubits,
            reps=self.reps,
            entanglement='linear',
            insert_barriers=True
        )
        return ansatz
    
    def create_chemistry_inspired_ansatz(self) -> QuantumCircuit:
        """
        Create chemistry-inspired ansatz circuit with mode-specific entanglement
        
        Returns:
            QuantumCircuit: Custom parameterized circuit
        """
        # Create parameter vector
        num_params = 2 * self.n_qubits * (self.reps + 1) + self.reps * 6  # Extra params for chemistry gates
        params = ParameterVector('', num_params)
        
        qc = QuantumCircuit(self.n_qubits)
        param_idx = 0
        
        # Initial layer of RY rotations (preparing ground vibrational state)
        for qubit in range(self.n_qubits):
            qc.ry(params[param_idx], qubit)
            param_idx += 1
        
        # Alternating layers of chemistry-inspired entanglement and rotations
        for rep in range(self.reps):
            # Chemistry-inspired entangling layer
            # NH2 group coupling (modes 0, 1, 5)
            if self.n_qubits > 5:
                qc.cx(0, 1)  # NH2 stretches
                qc.cx(1, 5)  # NH2 stretch to bend
                qc.cry(params[param_idx], 0, 5)  # Conditional coupling
                param_idx += 1
            
            # COOH group coupling (modes 2, 3)
            if self.n_qubits > 3:
                qc.cx(2, 3)  # COOH O-H to C=O
                qc.crz(params[param_idx], 2, 3)  # Phase coupling
                param_idx += 1
            
            # C-N backbone coupling (modes 3, 4)
            if self.n_qubits > 4:
                qc.cx(3, 4)  # C=O to C-N
                qc.cry(params[param_idx], 3, 4)  # Conditional coupling
                param_idx += 1
            
            # Inter-group coupling
            if self.n_qubits > 4:
                qc.cx(1, 4)  # NH2 to backbone
                qc.crz(params[param_idx], 1, 4)
                param_idx += 1
            
            # Linear entanglement for remaining modes
            for qubit in range(min(6, self.n_qubits - 1)):
                if qubit not in [1, 3, 4]:  # Skip already entangled qubits
                    next_qubit = (qubit + 1) % self.n_qubits
                    qc.cx(qubit, next_qubit)
            
            # Rotation layer
            for qubit in range(self.n_qubits):
                qc.ry(params[param_idx], qubit)
                param_idx += 1
                qc.rz(params[param_idx], qubit)
                param_idx += 1
        
        return qc
    
    def callback_function(self, eval_count, params, mean, std):
        """Callback function to track VQE progress"""
        self.iteration_count += 1
        self.energy_history.append(mean)
        
        if self.iteration_count % 10 == 0:
            energy_cm = mean / self.cm_to_hartree
            print(f"Iteration {self.iteration_count}: Energy = {mean:.6f} Ha ({energy_cm:.2f} cm^-1)")
    
    def run_classical_reference(self, hamiltonian: SparsePauliOp) -> Tuple[float, np.ndarray]:
        """
        Compute exact ground state for comparison
        
        Args:
            hamiltonian: Hamiltonian operator
            
        Returns:
            Ground state energy and eigenvector
        """
        print("Computing exact diagonalization reference...")
        # Convert to dense matrix for small systems
        H_matrix = hamiltonian.to_matrix()
        eigenvalues, eigenvectors = np.linalg.eigh(H_matrix)
        
        ground_energy = eigenvalues[0]
        ground_state = eigenvectors[:, 0]
        
        return ground_energy, ground_state
    
    def run_vqe_qiskit(self, optimizer_name: str = 'COBYLA', 
                      max_iterations: int = 200, 
                      use_chemistry_ansatz: bool = True) -> Dict[str, Any]:
        """
        Run VQE using Qiskit's built-in VQE algorithm
        
        Args:
            optimizer_name: Name of the optimizer ('COBYLA', 'SPSA', 'L_BFGS_B')
            max_iterations: Maximum number of iterations
            use_chemistry_ansatz: Whether to use chemistry-inspired ansatz
            
        Returns:
            Dictionary with VQE results
        """
        print("Setting up Qiskit VQE for NH2COOH vibrational ground state...")
        print(f"Number of qubits: {self.n_qubits}")
        print(f"Ansatz repetitions: {self.reps}")
        print("Vibrational modes:")
        mode_names = ["NH2 antisym stretch", "NH2 sym stretch", "COOH O-H stretch", 
                      "C=O stretch", "C-N stretch", "NH2 bending", "C-OH bending", "C-C-N bending"]
        
        for i, (freq, name) in enumerate(zip(self.frequencies[:self.n_qubits], mode_names[:self.n_qubits])):
            print(f"  Mode {i}: {freq:.0f} cm^-1 ({name})")
        
        # Create Hamiltonian
        hamiltonian = self.create_hamiltonian_pauli()
        print(f"Hamiltonian terms: {len(hamiltonian.paulis)}")
        
        # Classical reference
        exact_energy, exact_state = self.run_classical_reference(hamiltonian)
        print(f"Exact ground state energy: {exact_energy:.6f} Hartree")
        print(f"In cm^-1: {exact_energy/self.cm_to_hartree:.2f}")
        
        # Create ansatz
        if use_chemistry_ansatz:
            ansatz = self.create_chemistry_inspired_ansatz()
            print("Using chemistry-inspired ansatz")
        else:
            ansatz = self.create_ansatz_circuit()
            print("Using standard EfficientSU2 ansatz")
        print(f"Ansatz parameters: {ansatz.num_parameters}")
        
        # Choose optimizer
        optimizers = {
            'COBYLA': COBYLA(maxiter=max_iterations),
            'SPSA': SPSA(maxiter=max_iterations),
            'L_BFGS_B': L_BFGS_B(maxfun=max_iterations)
        }
        
        optimizer = optimizers.get(optimizer_name, SLSQP(maxiter=max_iterations))
        
        # Initialize VQE
        estimator = StatevectorEstimator()
        
        vqe = VQE(
            estimator=estimator,
            ansatz=ansatz,
            optimizer=optimizer,
            callback=self.callback_function
        )
        
        # Run VQE
        print(f"\nStarting VQE optimization with {optimizer_name}...")
        self.energy_history = []
        self.iteration_count = 0
        
        result = vqe.compute_minimum_eigenvalue(hamiltonian)
        
        # Extract results
        vqe_energy = result.eigenvalue
        optimal_params = result.optimal_parameters
        
        print(f"\nVQE Results:")
        print(f"VQE ground state energy: {vqe_energy:.6f} Hartree")
        print(f"In cm^-1: {vqe_energy/self.cm_to_hartree:.2f}")
        print(f"Error: {abs(vqe_energy - exact_energy):.6f} Hartree")
        print(f"Error in cm^-1: {abs(vqe_energy - exact_energy)/self.cm_to_hartree:.2f}")
        print(f"Relative error: {100*abs(vqe_energy - exact_energy)/abs(exact_energy):.4f}%")
        print(f"Optimizer evaluations: {result.cost_function_evals}")
        
        return {
            'vqe_energy': vqe_energy,
            'exact_energy': exact_energy,
            'error': abs(vqe_energy - exact_energy),
            'optimal_params': optimal_params,
            'result': result,
            'hamiltonian': hamiltonian,
            'ansatz': ansatz
        }
    
    def analyze_ground_state(self, optimal_params: Dict, ansatz: QuantumCircuit, 
                           hamiltonian: SparsePauliOp):
        """
        Analyze the VQE ground state
        
        Args:
            optimal_params: Optimal circuit parameters
            ansatz: The ansatz circuit
            hamiltonian: Hamiltonian operator
        """
        print("\nGround State Analysis:")
        
        # Create circuit with optimal parameters
        bound_circuit = ansatz.assign_parameters(optimal_params)
        
        # Get statevector
        statevector = Statevector.from_instruction(bound_circuit)
        state_array = statevector.data
        
        print("Significant state amplitudes (vibrational excitations):")
        mode_names = ["NH2-as", "NH2-s", "O-H", "C=O", "C-N", "NH2-bend"]
        
        for i, amplitude in enumerate(state_array):
            if abs(amplitude) > 0.05:  # Show states with >5% amplitude
                binary = format(i, f'0{self.n_qubits}b')
                prob = abs(amplitude)**2
                
                # Interpret the binary string as vibrational excitations
                excitations = []
                for j, bit in enumerate(binary):
                    if bit == '1':
                        mode_name = mode_names[j] if j < len(mode_names) else f"Mode{j}"
                        excitations.append(mode_name)
                
                excitation_str = ", ".join(excitations) if excitations else "Ground"
                print(f"|{binary}: {amplitude:.4f} (prob: {prob:.4f}) - {excitation_str}")
        
        # Calculate energy expectation value
        energy = statevector.expectation_value(hamiltonian)
        print(f"\nExpectation value H: {energy:.6f} Hartree ({energy/self.cm_to_hartree:.2f} cm^-1)")
        
        return statevector
    
    def plot_convergence(self):
        """Plot VQE convergence history"""
        if not self.energy_history:
            print("No convergence data available")
            return
            
        plt.figure(figsize=(15, 5))
        
        # Energy convergence in Hartree
        plt.subplot(1, 3, 1)
        plt.plot(self.energy_history, 'b-', linewidth=2, marker='o', markersize=3)
        plt.axhline(y=self.energy_history[-1], color='r', linestyle='--', 
                   label=f'Final: {self.energy_history[-1]:.6f} Ha')
        plt.xlabel('Iteration')
        plt.ylabel('Energy (Hartree)')
        plt.title('VQE Energy Convergence')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Energy in cm^-1
        plt.subplot(1, 3, 2)
        energy_cm = np.array(self.energy_history) / self.cm_to_hartree
        plt.plot(energy_cm, 'g-', linewidth=2, marker='s', markersize=3)
        plt.axhline(y=energy_cm[-1], color='r', linestyle='--',
                   label=f'Final: {energy_cm[-1]:.2f} cm')
        plt.xlabel('Iteration')
        plt.ylabel('Energy (cm)')
        plt.title('VQE Energy Convergence (cm)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Convergence rate
        plt.subplot(1, 3, 3)
        if len(self.energy_history) > 1:
            energy_diff = np.abs(np.diff(self.energy_history))
            plt.semilogy(energy_diff, 'r-', linewidth=2, marker='d', markersize=3)
            plt.xlabel('Iteration')
            plt.ylabel('|E| (Hartree)')
            plt.title('Energy Change per Iteration')
            plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def plot_vibrational_modes(self):
        """Plot the vibrational mode frequencies"""
        plt.figure(figsize=(12, 6))
        
        mode_names = ["NH2\nantisym", "NH2\nsym", "O-H\nstretch", "C=O\nstretch", 
                      "C-N\nstretch", "NH2\nbend", "C-OH\nbend", "C-C-N\nbend"]
        
        frequencies = self.frequencies[:self.n_qubits]
        colors = ['red', 'red', 'blue', 'green', 'green', 'orange', 'blue', 'purple']
        
        bars = plt.bar(range(len(frequencies)), frequencies, 
                       color=colors[:len(frequencies)], alpha=0.7, edgecolor='black')
        
        plt.xlabel('Vibrational Mode')
        plt.ylabel('Frequency (cm)')
        plt.title('NH2COOH Vibrational Mode Frequencies')
        plt.xticks(range(len(frequencies)), mode_names[:len(frequencies)], rotation=45)
        
        # Add frequency labels on bars
        for i, (bar, freq) in enumerate(zip(bars, frequencies)):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
                    f'{freq:.0f}', ha='center', va='bottom', fontweight='bold')
        
        plt.tight_layout()
        plt.show()

def main():
    """Main execution function"""
    print("NH2COOH (Carbamic Acid) Vibrational Ground State VQE with Qiskit")
    print("=" * 70)
    
    # Initialize VQE simulator with different system sizes
    n_qubits_list = [4, 6]  # Test different system sizes
    
    for n_qubits in n_qubits_list:
        print(f"\n{'='*20} System with {n_qubits} qubits {'='*20}")
        
        # Initialize Qiskit VQE simulator
        vqe_sim = NH2COOH_Qiskit_VQE(n_qubits=n_qubits, reps=2)
        
        # Plot vibrational modes
        vqe_sim.plot_vibrational_modes()
        
        # Run VQE with different optimizers
        optimizers = ['COBYLA', 'SPSA']
        results = {}
        
        for opt in optimizers:
            print(f"\n{'-'*15} {opt} Optimizer {'-'*15}")
            result = vqe_sim.run_vqe_qiskit(
                optimizer_name=opt, 
                max_iterations=200,
                use_chemistry_ansatz=True
            )
            results[opt] = result
            
            # Plot convergence
            vqe_sim.plot_convergence()
            
            # Analyze ground state
            vqe_sim.analyze_ground_state(
                result['optimal_params'], 
                result['ansatz'],
                result['hamiltonian']
            )
        
        # Compare optimizers
        print(f"\n{'-'*15} Optimizer Comparison {'-'*15}")
        plt.figure(figsize=(12, 8))
        
        # Energy comparison
        plt.subplot(2, 2, 1)
        opt_names = list(results.keys())
        opt_energies = [results[opt]['vqe_energy'] for opt in opt_names]
        exact_energy = results[opt_names[0]]['exact_energy']
        
        colors = ['blue', 'red', 'green', 'orange']
        bars = plt.bar(opt_names, opt_energies, color=colors[:len(opt_names)], alpha=0.7)
        plt.axhline(y=exact_energy, color='black', linestyle='--', label='Exact')
        plt.ylabel('Energy (Hartree)')
        plt.title('Final Energy Comparison')
        plt.legend()
        
        # Error comparison
        plt.subplot(2, 2, 2)
        opt_errors = [results[opt]['error'] for opt in opt_names]
        plt.bar(opt_names, opt_errors, color=colors[:len(opt_names)], alpha=0.7)
        plt.ylabel('Absolute Error (Hartree)')
        plt.title('Accuracy Comparison')
        plt.yscale('log')
        
        # Energy in cm^-1
        plt.subplot(2, 2, 3)
        opt_energies_cm = [energy/vqe_sim.cm_to_hartree for energy in opt_energies]
        exact_energy_cm = exact_energy/vqe_sim.cm_to_hartree
        
        bars = plt.bar(opt_names, opt_energies_cm, color=colors[:len(opt_names)], alpha=0.7)
        plt.axhline(y=exact_energy_cm, color='black', linestyle='--', label='Exact')
        plt.ylabel('Energy (cm)')
        plt.title('Energy Comparison (cm)')
        plt.legend()
        
        # Function evaluations
        plt.subplot(2, 2, 4)
        opt_evals = [results[opt]['result'].cost_function_evals for opt in opt_names]
        plt.bar(opt_names, opt_evals, color=colors[:len(opt_names)], alpha=0.7)
        plt.ylabel('Function Evaluations')
        plt.title('Computational Cost')
        
        plt.tight_layout()
        plt.show()
    
    print("\nSimulation completed successfully!")
    print("\nKey insights for NH2COOH vibrational ground state:")
    print("- VQE successfully captures ground vibrational state")
    print("- Chemistry-inspired ansatz shows improved performance")
    print("- Higher frequency modes (NH2, O-H stretches) dominate the spectrum")
    print("- Coupling between functional groups affects ground state energy")
    print("- Quantum advantage becomes apparent for larger molecular systems")

if __name__ == "__main__":
    main()
```
